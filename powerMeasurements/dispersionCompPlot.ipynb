```python
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fPath = r'F:\Data\jrowland\2018-10-23\Dispersion_comp'

voltage_arrays = []
GDDs = []

for file in os.listdir(fPath):
    if file.endswith("voltage_array.npy"):
        voltage_arrays.append(os.path.join(fPath, file))
        str_split = file.split('GDD')
        str_split2 = str_split[0].split('_')
        GDDs.append(int(str_split2[-2]))

GDDs
```

```python
x_axis = np.arange(0,1100,100)

for i,va in enumerate(voltage_arrays):
    power = np.load(va)
    plt.plot(x_axis,power*1000, label = 'Manual GDD {}'.format(GDDs[i]))
    plt.legend()
    plt.xlabel('PV power')
    plt.ylabel('Power on Sample (mW)')
sns.despine

```

```python
max_powers = [max(np.load(va)) for va in voltage_arrays]
max_powers = np.array(max_powers)
plt.plot(GDDs, max_powers*1000, '.')
plt.xlabel('Manual GDD')
plt.ylabel('Max Power on Sample (mW)')
```
